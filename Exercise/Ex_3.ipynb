```python
import sympy as sym
sym.init_printing()
from IPython.display import display, Math
```

```python
x,y = sym.symbols('x,y')
display(x)
```

```python
x**4
x/y
```

```python
sym.sqrt(2)
```

```python
display(Math('y \\times x^2'))
display(Math('\\sqrt{4u + 3z -45w} x'))
display(Math('x_{rr}'))
display(Math('\\frac{1+x}{2v - s^{4+3t}}'))
```

```python
display(Math('4x + 5y - 8z = 17'))
display(Math('\\sin(2\\pi ft + \\theta)'))
display(Math('e = mc^2'))
display(Math('\\frac{4 + 5x^2}{(1 + x)(1 - x)}'))
```

```python
mu,alpha,sigma = sym.symbols('mu,alpha,sigma')
expr = sym.exp((mu-alpha)**2 / (2*sigma**2))
display(expr)
```

```python
expr1 = x+4+2*y
expr1.subs({x:-4,y:3})
```

```python
expr2 = 3/x
display(Math(sym.latex(expr2)))
sym.latex(sym.sympify('3/4'))
```

```python
expr3 = x**2 + 4
for i in range(-2,3):
   display(Math('\\text {With x = %g} \\space : \\quad x^2 + 4 \\space \\qquad \\Rightarrow \\qquad %g^2 + 4 = %s' %(i,i, sym.latex(expr3.subs(x, i)))))
```

```python
x,y,z = sym.symbols('x,y,z')
ex= x**y * x**z
display(ex)
display(sym.simplify(ex))
```
