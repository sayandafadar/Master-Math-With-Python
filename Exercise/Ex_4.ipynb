```python
def check_for_one(num):
    return list(sym.factorint(num).values())

def count_factor(num):
    return len(list(sym.factorint(num).values()))

def check_for_factors(num):
    return list(sym.factorint(num).keys())
```

```python
for num in range(2,51):
    if count_factor(num) == 1 and check_for_one(num)[0] == 1:
        print(f'{num} is a prime number')
    else:
        print(f'{num} is a composite number with prime factors {check_for_factors(num)}')
```
