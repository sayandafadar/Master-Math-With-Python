```python
import sympy as sym
import numpy as np
from IPython.display import display, Math
```

```python
x =  sym.symbols('x')

expr = 6*x**2 + 7*x - 9
ans = sym.solve(expr)
display(Math('\\text{The solution to } %s \\space \\text{is} \\space x =  %s ' %(sym.latex(expr), sym.latex(ans))))
```

```python
y = sym.symbols('y')
expr = x/4 - x*y + 5
sym.solve(expr,y)
```

```python
q = sym.symbols('q')
expr = sym.Eq(3*q + 4/q + 3, 5*q + 1/q +1)
display(Math('\\text{Simplify:} \\quad %s'%(sym.latex(sym.simplify(expr)))))
display(Math('\\text{Ans:} \\quad q = %s' %(sym.latex(sym.solve(expr)))))
```

```python
expr = 2*q + 3*q**2 - 5/q - 4/q**3
display(sym.simplify(expr))
display(sym.cancel(expr))
```

```python
expr = (sym.sqrt(3) + sym.sqrt(15) * q) / (sym.sqrt(2) + sym.sqrt(10) * q)
display(expr)
display(Math(sym.latex(sym.simplify(expr))))
expr.subs(q,3).evalf()
```

```python
Fxy = (4+x)*(2-y)

mat = np.zeros([3,3])

for xi in range(0,3):
    for yi in range(0,3):
        mat[xi,yi] = Fxy.subs({x:xi,y:yi})
display(Math(sym.latex(sym.sympify(mat))))
```

```python
Fxy = x*y

table_range = range(0,10)
mat = np.zeros([len(table_range), len(table_range)]).astype(int)

for i in table_range:
    for j in table_range:
        mat[i,j] = Fxy.subs({x:i+1,y:j+1})
display(Math(sym.latex(sym.simplify(mat))))
```
