```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 1 
p = 1/(4*a)
h = 1
k = -2
n = 100
d = k-p

#x-axis points to evaluate the function
x = np.linspace(h-2,h+2,n)

#create the function
y = a*(x-h)**2 + k

#parabola
plt.plot(x,y, label='Parabola')

#vertex
plt.plot(h,k,'ro',label='Vertex')

#Focus
plt.plot(h,k+p,'go',label='Focus')

#Directrix
plt.plot(x[[0,-1]],[d,d], label='Directrix')

plt.legend()
plt.grid()
plt.show()
```

```python
x = np.linspace(-np.pi,2*np.pi,40)
y = np.linspace(0,4*np.pi,72)

X,Y = np.meshgrid(x,y)

Fxy = np.cos(X) + np.sin(Y)

plt.imshow(Fxy)
plt.show()
```

```python
X,Y = np.meshgrid(np.linspace(-2,2,100),np.linspace(-2,2,100))
s = 1

Z = np.exp(-(X**2+Y**2)/s)

plt.imshow(Z)
plt.axis('off')
plt.show()
```

```python
h = 2
k = -3
r = 3

axlim = r + np.max((abs(h),abs(k)))
x = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,x)

Fxy = (X-h)**2 + (Y-k)**2 - r**2

plt.contour(X,Y,Fxy,0)
plt.plot([-axlim,axlim],[0,0],'k--')
plt.plot([0,0],[-axlim,axlim],'k--')
plt.plot(h,k,'go')
plt.axis('square')
plt.show()
```

```python
h = [-1.5,1.5]
k = -3
r = [0.5,3]

x = np.linspace(-5,5,100)
X,Y = np.meshgrid(x,x)

for r in np.linspace(.5,3,15):
    for hi in h:
        Fxy = (X-hi)**2 + (Y)**2 - r**2
        plt.contour(X,Y,Fxy,0,colors=[r/3,r/3,r/3])


plt.axis('off')
plt.axis('square')
plt.plot(h,[0,0],'k',linewidth=3)
plt.savefig('venn.svg')
plt.show()
```

```python
h = 1
k = 2
a = 2
b = 3

axlim = np.max((a,b)) + np.max((abs(h), abs(k)))
x = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,x)

ellipse = (X-h)**2/a**2 + (Y-k)**2/b**2 - 1
plt.contour(X,Y,ellipse,0)
plt.plot(h,k,'go')
plt.axis('square')
plt.show()
```

```python
n = 16
h = 0
k = np.linspace(-4,4,n)
a = abs(k)
b = 4

x = np.linspace(-8,8,100)
X,Y = np.meshgrid(x,x)

for i in range(0,n):
    ellipse = (X-h)**2/a[i]**2 + (Y-k[i])**2/b**2 - 1
    plt.contour(X,Y,ellipse,0,colors=[i/19,0,i/n])


plt.axis('square')
plt.axis('off')
plt.show()
```

```python
n = 16
ab = np.linspace(1,5,n)

axlim = 8
x = np.linspace(-axlim,axlim,100)
X,Y = np.meshgrid(x,x)

for i in range(0,n):
    F1 = X**2/ab[i]**2 - Y**2/ab[i]**2 - 1
    plt.contour(F1,0,colors=[i/n,0,i/n])
    F2 = Y**2/ab[i]**2 - X**2/ab[i]**2 - 1
    plt.contour(F2,0,colors=[0,i/n,i/n])

plt.axis('off')
plt.axis('square')
plt.savefig('bayblade.svg')
plt.show()
```
