```python
import matplotlib.pyplot as plt
import sympy as sym
import numpy as np
import sympy.plotting.plot as symplot
from matplotlib.patches import Polygon
```

```python
x = 3
y = 5

plt.plot(x,y,'ro')
plt.axis('square')
plt.axis([-6,6,-6,6])
plt.grid()
plt.show()
```

```python
x = [-4,5,2,4,9,8,3]
y = [-7,0,7,3,9,2,1]

for i in range(0,len(x)):
    plt.plot(x[i],y[i],'o',label='Point %s' %(1+i))

plt.axis('square')
plt.grid()
plt.legend()

plt.show()
```

```python
x,y = sym.symbols('x,y')
y = x**2 - 3*x

x_range = range(-10,11)

for p in x_range:
    plt.plot(p, y.subs({x:p}), 'o')

plt.xlabel('x')
plt.ylabel('f(x) = %s' %sym.sympify(y))
plt.show()
```

```python
x = [-5,5]
m = -.3
b = .1

y = m*np.array(x)+b

plt.plot(x,y,label='y=%sx+%s' %(m,b))

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)

axis = plt.gca()
plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')

plt.legend()

plt.show()
```

```python
x = [-5,5]
m = [.7,-1.25,.45]
b = [-2,0.75,.65]

for i in range(0,len(m)):
    y = m[i]*np.array(x)+b[i]
    plt.plot(x,y,label='y=%sx+%s' %(m[i],b[i]))

plt.axis('square')
plt.grid()
plt.xlim(x)
plt.ylim(x)
plt.xlabel('x')
plt.ylabel('y')
plt.title('The Plot')

axis = plt.gca()
plt.plot(axis.get_xlim(), [0,0], 'k--')
plt.plot([0,0], axis.get_ylim(), 'k--')

plt.legend()

plt.show()
```

```python
x = np.linspace(-4,4,200)

for i in range(-1,4):
    y = x**i
    plt.plot(x,y,label='$y=x^%s$' %(i),linewidth=4)

plt.legend()
plt.xlim([x[0],x[-1]])
plt.ylim([-20,20])
plt.show()
```

```python
x,y = sym.symbols('x,y')

y = x**2

p = symplot(y,show=False)
p.xlim = [0,50]
p[0].line_color = 'r'
p.title = 'This is a sympy plot!'
p.show()
```

```python
x,a = sym.symbols('x,a')

expr = a/x
p = symplot(expr.subs(a,1),(x,-5,5),show=False)
p.ylim = [-5,5]
p[0].label = '$f(x) = %s$' %sym.latex(expr.subs(a,1))
p2 = symplot(expr.subs(a,2),show=False)
p.extend(p2)
p[0].line_color = 'r'
p[1].label = '$f(x) = %s$' %sym.latex(expr.subs(a,2))
p.ylim = [-5,5]
p.legend = True
p.show()
```

```python
a,x = sym.symbols('a,x')

y = a/(x**2 - a) 
p = symplot(y.subs(a,1),(x,-4,4),show=False)
p[0].label = '$f(x) = %s$' %sym.latex(y.subs(a,1))
for num in range(2,5):
    p2 = symplot(y.subs(a,num),(x,-4,4),show=False)
    p.extend(p2)
    p[num-1].line_color = list(np.random.rand(3))
    p[num-1].label = '$f(x) = %s$' %sym.latex(y.subs(a,num))
    
p.legend = True
p.ylim = [-10,10]
p.show()
```

```python
A = np.zeros((10,14))

for i in range(0, np.shape(A)[0]):
    for j in range(0, np.shape(A)[1]):
        A[i,j] = 3*i - 4*j

plt.imshow(A)

for i in range(0, np.shape(A)[0]):
    for j in range(0, np.shape(A)[1]):
        plt.text(j,i,int(A[i,j]), horizontalalignment='center', verticalalignment='center',fontdict={'color' : 'black'})

plt.set_cmap('twilight_shifted')
plt.savefig('ThisIsANiceFigure.svg')
plt.show()
```

```python
N = 10
C = np.zeros((N,N))

for i in range(0, np.shape(C)[0]):
    for j in range(0, np.shape(C)[1]):
        C[i,j] = (-1)**(i+j)

plt.set_cmap('gray')
plt.tick_params(labelleft=False,labelbottom=False)
plt.imshow(C)
plt.show()
```

```python
y = np.array([[1,1],[2,3],[3,1]])
p = Polygon(y,facecolor='m',alpha=.3)

y1 = np.array([[2,1],[3,2],[2,3]]) 
p1 = Polygon(y1,facecolor='b',alpha=.3)

fig,ax = plt.subplots()

ax.add_patch(p)
ax.add_patch(p1)
ax.set_ylim([0,4])
ax.set_xlim([0,4])
plt.show()
```

```python
x = np.linspace(-2,2,101)
f = -x**2

pnts = np.vstack((x,f)).T
p = Polygon(pnts,facecolor='g',alpha=.3)

y = np.array([[-.5,-4],[-.5,-2.5],[.5,-2.5],[.5,-4]])
p1 = Polygon(y,facecolor='k')
fig,ax = plt.subplots()
ax.add_patch(p)
ax.add_patch(p1)
plt.plot(x,f,'k')
plt.plot([-2,2],[-4,-4],'k')
plt.axis('off')
plt.show()
```
