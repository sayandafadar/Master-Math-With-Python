```python
def exponent(x,y):
    return x**y

def fraction(x,y):
    return x/y

def compute():
    num_1 = int(input("input X: "))
    num_2 = int(input("input Y: "))
    choice = int(input('Press "1" to compute or "2" to compute ' %()))

    if choice == 1:
        print(exponent(num_1, num_2))
    else: 
        print(fraction(num_1, num_2))


```

```python
import random
def guessTheNumber():
    correct_number = random.randint(1, 101)
    user_input = int(input("Guess a number between 1 and 100: "))
    game_on = True

    while game_on:
        if user_input > correct_number:
            print("Guess higher!")
            user_input = int(input("Guess again "))
        elif user_input < correct_number:
            print("Guess lower!")
            user_input = int(input("Guess again "))
        else:
            print(f"Got it! The right number was {correct_number} and your final guess was {user_input}")
            game_on = False
            break
```

```python
guessTheNumber()
```
