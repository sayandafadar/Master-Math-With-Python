```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
t = np.linspace(-6*np.pi, 6*np.pi, 1000)

a = t

x = a * np.cos(t)**3
y = a * np.sin(t)**3

plt.plot(x,y)
plt.show()
```

```python
t = np.linspace(-6*np.pi, 6*np.pi, 1000)

fig, axs = plt.subplots(3,3,figsize=(10,6))

row, col = np.indices((3,3))

for n in range(0,9):
    a = t**n

    x = a * np.cos(t)**3
    y = a * np.sin(t)**3

    r = row.ravel()[n]
    c = col.ravel()[n]

    axs[r, c].plot(x,y,'k')
    axs[r, c].axis("square")
    axs[r, c].axis("off")

plt.show()
```

```python
k = 4.5

t = np.linspace(0,8*np.pi,1000)

x = np.cos(k*t) * np.cos(t)
y = np.cos(k*t) * np.sin(t)

plt.plot(x,y,'m',linewidth=3)
plt.axis('square')
plt.axis('off')
plt.show()
```

```python
t = np.linspace(0,8*np.pi,1000)
n = np.linspace(0,4,9)

fig, axs = plt.subplots(3,3,figsize=(10,6))
row, col = np.indices((3,3))

for k in range(9):
    x = np.cos(n[k]*t) * np.cos(t)
    y = np.cos(n[k]*t) * np.sin(t)

    r = row.ravel()[k]
    c = col.ravel()[k]

    axs[r, c].plot(x,y,'k')
    axs[r, c].set_title('k=%s' %n[k])
    axs[r, c].axis("square")
    axs[r, c].axis("off")

plt.show()
```
